```python
# Bibliotecas
import speech_recognition as sr
from gtts import gTTS
import os
import playsound
import pyjokes
import webbrowser

#Módulo de Fala (Text-to-Speech) ---
def falar(texto):
    """Converte um texto para áudio e o reproduz."""
    print(f"Assistente: {texto}")
    try:
        tts = gTTS(text=texto, lang='pt-br')
        arquivo_audio = "resposta.mp3"
        tts.save(arquivo_audio)
        playsound.playsound(arquivo_audio)
        os.remove(arquivo_audio)
    except Exception as e:
        print(f"Ocorreu um erro na fala: {e}")

# Módulo de Audição (Speech-to-Text) ---
def ouvir_microfone():
    """Captura o áudio do microfone e o converte para texto."""
    microfone = sr.Recognizer()
    with sr.Microphone() as source:
        microfone.adjust_for_ambient_noise(source)
        print("Diga 'YouTube' ou 'Me conte uma piada'...")
        audio = microfone.listen(source)
    try:
        frase = microfone.recognize_google(audio, language='pt-BR')
        print(f"Você disse: {frase}")
        return frase.lower()
    except sr.UnknownValueError:
        return "" # Retorna vazio se não entender
    except sr.RequestError:
        falar("Desculpe, estou com problemas no meu serviço de reconhecimento de voz.")
        return ""

# Modulo de Comandos e Respostas ---
def responder(texto_comando):
    """Executa ações com base no comando de voz recebido."""
    if 'youtube' in texto_comando:
        falar("O que você gostaria de pesquisar no YouTube?")
        termo_pesquisa = ouvir_microfone()
        if termo_pesquisa:
            url = f"https://www.youtube.com/results?search_query={termo_pesquisa}"
            webbrowser.get().open(url)
            falar(f"Aqui estão os resultados para {termo_pesquisa} no YouTube.")

    elif 'piada' in texto_comando:
        piada = pyjokes.get_joke(language='pt', category='all')
        falar(piada)

    elif 'sair' in texto_comando or 'desligar' in texto_comando:
        falar("Até logo!")
        return True

    return False

# Loop Principal
if __name__ == "__main__":
    falar("Olá! Estou pronto para obedecer.")
    while True:
        comando = ouvir_microfone()
        if comando:
            parar = responder(comando)
            if parar:
                break
```
